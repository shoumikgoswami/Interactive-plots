```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option("display.max_columns",None) 
pd.set_option("display.max_rows",None) 
import warnings
warnings.filterwarnings("ignore")
from IPython.display import Image
sns.set(style="darkgrid", palette="pastel", color_codes=True)
sns.set_context("paper")
from datetime import datetime
import datapane as dp
```

```python
df = pd.read_csv('./archive/covid_19_clean_complete.csv')
```

```python
df.head()
```

```python
#Renaming the coulmns for easy usage
df.rename(columns={'Date': 'date', 
                     'Province/State':'state',
                     'Country/Region':'country',
                     'Lat':'lat', 'Long':'long',
                     'Confirmed': 'confirmed',
                     'Deaths':'deaths',
                     'Recovered':'recovered'
                    }, inplace=True)

# Active Case = confirmed - deaths - recovered
df['active'] = df['confirmed'] - df['deaths'] - df['recovered']
```

```python
df['date'] = pd.to_datetime(df['date'])
country_list = list(df['country'].unique())
```

```python
plot = plt.plot(df['date'][df['country']=='India'], df['confirmed'][df['country']=='India'])
plt.show()
fig1 = plot
```

```python
dp.Report(
    dp.Plot(plot))
```

```python
plot_list = []

plt.figure(figsize=(10, 5), dpi=300)

for country in country_list:
    plot = dp.Plot(df[df['country']==country].plot.scatter(x='date', y='confirmed'), label=country)
    plot_list.append(plot)
```

```python
report = dp.Report(
    dp.Text('''## Dropdown using Datapane'''),
    dp.Select(blocks = plot_list)
)
```
